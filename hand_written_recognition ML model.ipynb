```python
#importing necessary modules
import tensorflow as tf
import numpy as np
import gradio as gr
from PIL import Image as PILImage
```

```python
#loading the mnist dataset

mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels)=  mnist.load_data()
```

```python
#Normalize the image to range[0,1]
train_images, test_images = train_images/255.0,test_images/255.0
```

```python
#Create a Neural Network model
model= tf.keras.Sequential([
     tf.keras.layers.Flatten(input_shape=(28,28)), #flatten the images to 1D array
     tf.keras.layers.Dense(128,activation='relu'), #Hidden layer with Relu activation
     tf.keras.layers.Dropout(0.2),#Dropout layer to prevent overfitting
     tf.keras.layers.Dense(10,activation='softmax') #output layer with 10 units (for digits 0-9)
])
```

```python
model.compile(optimizer = 'adam',
              loss= 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training the model
model.fit(train_images,train_labels, epochs=5)

```

```python
def predict_digit(image):

    if image is None:
        return 'No image provided'
    # Convert the Gradio image to a format compatible with TensorFlow
    image = np.array(image)
    
    # Resize the image to 28x28 if it's not already in the correct size
    image = PILImage.fromarray(image).resize((28, 28))
    image = np.array(image)
    
    # Convert the image to grayscale if it's in color (3 channels)
    if len(image.shape) == 3:
        image = np.dot(image[..., :3], [0.299, 0.587, 0.114])  # RGB to grayscale
    
    # Normalize the image to [0, 1]
    image = image / 255.0
    
    # Reshape to add a batch dimension
    image = np.expand_dims(image, axis=(0,-1))
    
    # Make prediction
    prediction = model.predict(image)
    digit = np.argmax(prediction)  # Get the predicted digit
    return str(digit)  # Return as string for Gradio's textbox
```

```python
# Create the Gradio interface with drawing capability
interface = gr.Interface(
    fn=predict_digit,
    inputs=gr.Sketchpad(),
    outputs=gr.Textbox(),
    live=True)
```

```python
# Launch the interface
interface.launch()
```
